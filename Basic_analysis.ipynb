```python
def CPUtemp_vs_time(num, CPU_table_KID):
    A=[]
    l=len(CPU_table_KID[0])

    for i in range(1,l+1):
        A.append(i)

    plt.plot(A,CPU_table_KID[num])
    plt.xlabel("time[s]") 
    plt.ylabel("CPU_temperature")
    plt.title('CPU_Temperature vs time')
    plt.legend()
    plt.show()

def FAN_vs_time(num, FAN_table_KID):
    A=[]
    l=len(FAN_table_KID[0])
    for i in range(1,l+1):
        A.append(i)

    plt.plot(A,FAN_table_KID[num])
    plt.xlabel("time[s]") 
    plt.ylabel("fan rotation speed")
    plt.title('FAN vs time')
    plt.legend()
    plt.show()
```
